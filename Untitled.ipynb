```python
file = open("data/zipcodes.csv","r")
```

```python
lines = file.read()
```

```python
lines = lines.split("\n")
```

```python
ziptable = {}
```

```python
for line in range(1,len(lines)):
    thisline = lines[line].split(',')
    zip = thisline[0]
    coords = thisline[1:3]
    ziptable[zip] = coords
```

```python
ziptable
```

```python
f = open("ziptable.txt","w")
f.write( str(ziptable) )
f.close()
```
